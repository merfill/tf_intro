```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

w0 = -2.3
w1 = 1.5

np.random.seed(42)
mixed = .2 * np.random.randn(100)
plt.scatter(range(len(mixed)), mixed)
```

```python
X = np.arange(1., 5., .01)
Y_r = w0 + w1 * X
Y = []
for x in X:
    Y.append(w0 + w1 * x + np.random.choice(mixed))

plt.scatter(X, Y, c='r', s=2)
plt.plot(X, Y_r, 'b')
```

```python

def calc_grad(x, w0, w1, yh):
    y = w0 + x * w1 - yh
    return [y, y * x]


def calc_delta(X, w0, w1, Yh):
    delta = [0., 0.]
    for i in range(len(X)):
        d = calc_grad(X[i], w0, w1, Yh[i])
        delta[0] += d[0]
        delta[1] += d[1]
    return [delta[0] / len(X), delta[1] / len(X)]
    
alpha = .01
w = np.random.randn(2)


W1 = []
W0 = []
W0.append(w[0])
W1.append(w[1])
for i in range(10000):
    delta = calc_delta(X, w[0], w[1], Y)
    w[0] = w[0] - alpha * delta[0]
    w[1] = w[1] - alpha * delta[1]
    W0.append(w[0])
    W1.append(w[1])
```

```python
def plot_step(step):
    plt.figure(2)
    plt.scatter(X, W0[step] + W1[step] * X, c='r', s=2)
    plt.plot(X, Y_r, 'b')

def plot_weights(W0, W1):
    plt.figure(1)
    plt.scatter(W0, W1, c='g', s=20)
```

```python
plot_weights(W0[:1], W1[:1])
plot_step(0)
```
